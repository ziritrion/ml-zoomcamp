```python
import torch

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
# Images
imgs = ['data/cars/training_images/vid_4_600.jpg']  # batch of images
```

```python
# Inference
results = model(imgs)
```

```python
# Results
results.print()
```

```python
#results.xyxy[0]  # img1 predictions (tensor)
results.pandas().xyxy[0]  # img1 predictions (pandas)
```
